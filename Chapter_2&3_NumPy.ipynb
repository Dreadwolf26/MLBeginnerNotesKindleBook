reshape() method

reshape gives numpy arrays a new shape without changing any of the data within the array

Suppose we have a 1D array with 8 elements and we want to reshape it to a (4,2) 2D array

It is essential to now how to reshape 1D arrays to 2D because some of the modules in Scikit-Learn require 2D arrays as input

```python
arr2 = np.array([1,2,3,4,5,6,7,8])
#When calling reshape below to reshape the array you are creating 4 rows with
#2 columns
#you can observe the printed outputs below
reshaped_array = arr2.reshape((4,2))
print(arr2)
print(reshaped_array)
```

A pandas series is a 1D (one dimensional) Data structure.

A DataFrame is a 2D (two dimensional) labeled data structure.

```python
import pandas as pd

list1 = [1,2,3,4,5]
print(list1, end='\n\n')
#this will print the series in a vertical line
series1 = pd.Series(list1)
print(series1, end='\n\n')
#this will print the series in a vertical line
series2 = pd.Series(list1, index=['P', 'Q', 'R', 'S', 'T'])
print(series2, end='\n\n')
#combining the above will list each series in a vertical line next to each other

#The column on the left will be the index of the series
```

# corr() method

This method gives us the pairwise correlation coefficients.

I needed an explanation of pairewise correlation coefficients and GPT delivered:

Pairwise correlation coefficients are a way to measure how two variables change together. In the context of machine learning, it helps us understand how two different features (or variables) in our dataset are related to each other.

Imagine you have a dataset with two features: "Hours of study" and "Exam score." Pairwise correlation coefficient tells you whether there's a relationship between these two features, and if so, how strong it is.

Here's a simple breakdown:

1. **What it measures**: Pairwise correlation coefficient measures the strength and direction of the relationship between two variables.

2. **Values it can take**: The value of a correlation coefficient ranges between -1 and 1.
   - A correlation of 1 means a perfect positive relationship: as one variable increases, the other also increases.
   - A correlation of -1 means a perfect negative relationship: as one variable increases, the other decreases.
   - A correlation of 0 means there's no linear relationship between the variables.

3. **How it's calculated**: It's calculated using a formula that essentially measures how much the variables deviate from their respective means together.

4. **Interpreting the coefficient**:
   - A positive coefficient means the variables tend to move in the same direction: when one increases, the other tends to increase.
   - A negative coefficient means the variables tend to move in opposite directions: when one increases, the other tends to decrease.
   - The closer the coefficient is to 1 or -1, the stronger the relationship. A coefficient close to 0 indicates a weak or no relationship.

5. **Significance**: It's important to note that correlation doesn't necessarily imply causation. Just because two variables are correlated doesn't mean that changes in one cause changes in the other.

In summary, pairwise correlation coefficients help us understand how features in our dataset are related to each other, which can be useful in various machine learning tasks like feature selection, understanding dependencies, and preprocessing data.

See code examples below:

See docstring within code this will answer why you need to drop a column if the correlation is too high:

If the correlation between two columns (features) in your dataset is too high, it means that these two columns are providing very similar information. Having highly correlated features in your dataset can lead to several issues in machine learning models:

1. **Redundancy**: Highly correlated features essentially carry the same information. Having redundant features doesn't provide additional useful information to the model but can increase the complexity and computational cost.

2. **Overfitting**: When you have highly correlated features, your model might overweight the importance of these features, leading to overfitting. Overfitting occurs when a model learns the training data too well, including noise or irrelevant patterns, which results in poor generalization to unseen data.

3. **Interpretability**: In some cases, highly correlated features can make it difficult to interpret the model. It becomes unclear which feature is truly influencing the outcome, as both features provide similar information.

4. **Model Performance**: Highly correlated features can lead to instability in model performance. Small changes in the data or noise can cause large variations in the model's predictions.

By dropping one of the highly correlated columns, you can mitigate these issues. It helps to simplify the model, reduce overfitting, and improve interpretability. However, it's essential to choose which column to drop carefully, considering domain knowledge, the importance of the features, and the specific requirements of the machine learning task. Typically, you might keep the feature that is more interpretable or has more significance in the context of the problem you're solving.
