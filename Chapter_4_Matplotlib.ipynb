# 4.2.4 Plotting a line graph



The plot() function joins the points in the graph based on the order of the elements in the first lsit (which represent x values)

So plot starts at x = 7 then x = 1 all the way through the list.

The Y-axis is of course display to the left of the graph and its corresponding values in the l_y variable.

The first code shows the line chart that is impossible to read and is plotted incorrect.

```python
l_x = [7,1,4,8,5,2,3]
l_y = [98,2,32,128,15,28,18]
plt.plot(l_x,l_y)
```

To plot the graph correct see code example 2 and I will explain here as well.

We pass in the l_x and l_y variables into python zip() function This pairs the corresponding element in each list into tuples and returns a zip object, which we then assign to zipped. We then need to sort the data in ascending order using the sorted() function and assign it to the variable sorted_zip.Since sorted has now taken place and the data is correct we need to 'unzip' the zip object by calling the zip() function and passing in an * before the variable name sorted_zip. Now if you print l_x and l_y you will get both of the list in a sorted fashin in ascending order. We then pass l_x and l_y into plot() and the line graph shows correctly

```python
zipped = zip(l_x,l_y)
sorted_zip = sorted(zipped)
l_x, l_y = zip(*sorted_zip)
#showing sorted lists
print(l_x)
print(l_y)
#plotting sorted lists
plt.plot(l_x,l_y)
```

 # 4.4 Designing out Charts

 We are working in matplotlib for charact customization essentially the same parameters will be used in pandas.

 for color you will have to pass in the color parameter in the function call. Example color = 'red'

 For Scatter plaots (scatter()) you will specify the marker parameter. Example marker = 'x' Here is a list of acceptable markers: https://matplotlib.org/stable/api/markers_api.html

 To set the size of the marker we will use the s parameter for scatter plots ms for line graphs. numbers will be assigned to these parameters.

 Besides the ones mentioned above we can specify labels for the data using the label parameter. This is useful if we are wanting to plot multiple sets of data within the same chart.

 We can set the title, x and y axis and tick values on our axis using title(), xlabel(), ylabel(), xticks(), yticks()
 The first three will accept strings as arguments and the last 2 will accept numbers

 We can also add legend() and grid() (for gridlines)

 We can also specify our chart size with the figure() fundtion to create a figure option. We pass in the figsize parameter to set the width and heign of the figure in inches.

```python
x1 = [1,2,4,6,8,9,10]
y1 = [3,4,6,12,4,5,7]
y2 = [5,1,2,6,8,9,1]
#establishing the size of the chart
plt.figure(figsize=(10,5))
#Scatterplot specifications
plt.scatter(x1,y1,color='black', s=50, marker='x', label='With Reward')
#y2 scatter plot specification
plt.scatter(x1, y2, color='darkgray', s=40, marker='o', label='Without reward')
#setting the legend
#loc is location
plt.legend(loc='best')
#setting gridlines
plt.grid()
#specifying tick values
plt.xticks([0,2,4,6,8,10])
#specifying y ticks
plt.yticks(range(0,13))
#label for x
plt.xlabel('Age')
#label for y
plt.ylabel("Number of Tries")
#title of chart
plt.title('Designing our charts')
```

# The object oriented interface

So far we have only covered using the pyplot interface while this is easy to use and convenient it can be restrictive with lesss control over the charts. for instance plotting multiple sets of data, it is harder to state wheter we want to plot data on one chart or on multiple seperate charts.

In this sections we will discuss object-oriented interface just like the title says. This offers more flexibility and control over the layout of the charts.

To undersatnad how this works we must first discuss the anatomy of a figure in Matplotlib. There are two important objests in matplotlib the Figure and the Axes objects.

A figure object (figure) is a top level container of a plot. Within each figure, we can have multiple Axes objects. you can think if Axes objects as an individual plot inside of a figure option.

To work with figure and axes objects you first need to create them.

The easiest way to do this is use the subplots() function this function returns a figure object and a set of axes.

We typically pass the number of rows and columns of the subplot grid to the function. If we do not specify the number of rows and columns the default will be 1. We also pass in the figsize parameter to the function to set the figure size.

```python
#fig has to be passed into the variable name to specify fig size
fig, my_ax = plt.subplots()
a = [1,2,3,4]
b = [7,3,1,4]
my_ax.scatter(a,b)
```

If wanting to plot in the same manner using Pandas plot()

```python
#pass in fig for size and variable name
fig, my_ax2 = plt.subplots()
df = pd.DataFrame ({'A':[1,2,3,4], 'B': [7,3,1,4]})
df.plot(kind='scatter', x='A', y='B', ax=my_ax2)
```

Plotting two charts on a single figure

```python
fig, axs = plt.subplots(1,2, figsize=(10,5))
df = pd.DataFrame({'A':[1,2,3,4], 'B':[4,2,3,1]})
#The numbers in the first bracket are like a map
axs[0].scatter(df['A'], df['B'])
axs[1].plot(df['A'], df['B'])
axs[0].set_title('Chart 1')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Number of tries')
axs[0].set_xticks([1,2,3,4])
axs[0].set_yticks([1,2,3,4])

```
